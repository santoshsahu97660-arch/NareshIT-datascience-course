## FORMAL ARGUMENT & ACTUAL ARGUMENT 

```python
def person(name,age):
    print(name)
    print(age)
    
person('nit',23,34)
```

```python
def person(name,age):
    print(name)
    print(age)
    
person('nit',23)
```

# keyword argument

```python
def person(name,age):
    print(name)
    print(age+1)
    
person(age=23, name='nit')
```

```python
def person(name,age):
    print(name)
    print(age+1)
    
person(age1=23, name='nit')
```

```python
def person(name,age1):
    print(name)
    print(age1+1)
    
person(age1=23, name='nit')
```

```python
def person(name,age,city):
    print(name)
    print(age+1)
    print(city)
    
person(age=23, name='nit', city = 'hyd')
```

```python
def person(name, age=18): 
    print(name)
    print(age)
    
person('nit', 24)
```
